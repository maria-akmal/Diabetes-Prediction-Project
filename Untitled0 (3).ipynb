```python

df['polyuria']=df['polyuria'].astype(str)
df['polydipsia']=df['polydipsia'].astype(str)
df['delayed_healing']=df['delayed_healing'].astype(str)

sns.set(style="whitegrid" , palette="pastel")
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
for col in df.columns:
    if df[col].dtype == 'object':
        df[col] = le.fit_transform(df[col])

print("تم تحويل البيانات للغة الأرقام!")
df.head()
```

```python
X = df.drop('class', axis=1)
y = df['class']
```

```python
from sklearn.preprocessing import StandardScaler

Scaler = StandardScaler()
X = Scaler.fit_transform(X)
```

# Exploratory Data Analysis (EDA)

```python
sns.histplot(df['age'], kde=True)
plt.title("Age Distribution")
plt.show()
```

# **Modeling and Machine Learning Implementation**

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=59
)
```

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(random_state=42)
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score

print(f"Final Project Accuracy: {accuracy_score(y_test, y_pred) * 100:.2f}%")
```

## **Model Evaluation**( Present findings in a well-structured report)

```python
from sklearn.metrics import classification_report

print("Classification Report:")
print(classification_report(y_test, y_pred))
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns

cm = confusion_matrix(y_test, y_pred)

plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```

```python
from sklearn.model_selection import cross_val_score

scores = cross_val_score(model, X, y, cv=5)
print("CV Accuracy:", scores.mean())
```

# Feature Importance

```python
importances = model.feature_importances_
feature_names = df.drop('class', axis=1).columns

feature_importance_df = pd.DataFrame({
    'Feature': feature_names,
    'Importance': importances
})

feature_importance_df = feature_importance_df.sort_values(
    by='Importance', ascending=False
)
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(
    x='Importance',
    y='Feature',
    data=feature_importance_df.head(5)
)
plt.title('Top 5 Symptoms Influencing Prediction')
plt.show()
```
